```python
import numpy as np
```

1. Criando um kernel gaussiano

```python
size = 3 
sigma=1.0
ax = np.arange(-(size // 2), size // 2 + 1)
print("ax: ")
print(ax)
xx, yy = np.meshgrid(ax, ax)
print("xx: ")
print(xx)
print("yy:")
print(yy)
kernel = np.exp(-(xx**2 + yy**2) / (2 * sigma**2))
print("kernel: ")
print(kernel)
print("kernel sum: ", kernel.sum())
kernel /= kernel.sum()
print("normalized kernel: ")
print(kernel)
print("kernel sum: ", kernel.sum())
```

2. Vamos aplicar o kernel a uma imagem exemplo

```python
# Cria uma matriz 5x5 simples
image = np.array([
    [1, 2, 3, 2, 1],
    [2, 4, 6, 4, 2],
    [3, 6, 9, 6, 3],
    [2, 4, 6, 4, 2],
    [1, 2, 3, 2, 1]
], dtype=float)
```

```python
img_height, img_width = image.shape
print("Image height: ", img_height)
print("Image width: ", img_width)
k_height, k_width = kernel.shape
print("Kernel height: ", k_height)
print("Kernel width: ", k_width)
```

```python
# Calcula padding necessário
pad_h = k_height // 2
pad_w = k_width // 2
print("Padding height: ", pad_h)
print("Padding width: ", pad_w)
```

```python
# Aplica padding por replicação de bordas
padded_image = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='edge')
print("Padded image: ")
print(padded_image)
```

```python
# Inicializa imagem de saída
output = np.zeros_like(image)

# Aplica convolução
for i in range(img_height):
    for j in range(img_width):
        # Extrai região
        region = padded_image[i:i+k_height, j:j+k_width]
        # Multiplica elemento a elemento e soma
        output[i, j] = np.sum(region * kernel)
    
```
